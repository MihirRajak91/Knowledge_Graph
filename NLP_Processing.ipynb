```python
import pandas as pd
import numpy as np
import nltk
from nltk.tokenize import sent_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize
nltk.download('punkt')
nltk.download('stopwords')
```

```python
df = pd.read_csv('OnlineKhabar_Final_Data.csv')
```

```python
def process(text):
    sentences = sent_tokenize(text)
    
    stop_words = set(stopwords.words('english'))
    
    sentences = [' '.join([word for word in word_tokenize(sentence) if word.lower() not in stop_words]) for sentence in sentences]
    
    ps = PorterStemmer()
    
    sentences = [' '.join([ps.stem(word) for word in word_tokenize(sentence)]) for sentence in sentences]
    
    return sentences
```

```python
df['Title_Processed'] = df['Title'].apply(process)

df['Article_Processed'] = df['Article'].apply(process)
```
