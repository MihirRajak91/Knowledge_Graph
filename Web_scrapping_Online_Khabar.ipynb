## Scrapping News From OnlineKhabar

```python
# importing libraies
from bs4 import BeautifulSoup as soup
import requests
```

```python
def news_getter(num):
    url = 'https://english.onlinekhabar.com/category/sports/page/{}'.format(num)
    r = requests.get(url)
    b = soup(r.content, 'lxml')
    
    # to get title
    title = get_title(b)
    
    # to get link
    links = get_link(b)
    
    # to get artice 
    article = get_article(links)
    
    return title, links, article
```

```python
def get_title(b):
    title = [news.text for news in b.findAll('h2')]
    return title
```

```python
def get_link(b):
    links = [news.a['href'] for news in b.findAll('h2') if news.a]
    return links
```

```python
def get_article(links):
    article = []
    for link in links:
        page = requests.get(link)
        bsobj = soup(page.content)
        for news in bsobj.findAll('div', {'class': 'post-content-wrap'}):
            print(news.text.strip())
            article.append(news.text.strip())
    return article
```

```python
all_titles = []
all_links = []
all_articles = []
```

```python
for i in range(1, 11):
    titles, links, articles = news_getter(i)
    
    all_titles.extend(titles)
    all_links.extend(links)
    all_articles.extend(articles)
```

```python
print(len(all_titles))
print(len(all_links))
print(len(all_articles))
```

```python
import pandas as pd
all_titles_subset = all_titles[:315]
all_articles_subset = all_articles[:315]

df = pd.DataFrame({
    'Title': all_titles_subset,
    'Article': all_articles_subset
})
```
